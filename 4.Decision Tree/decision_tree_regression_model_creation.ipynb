```python
import pandas as pd #import pandas library to handle file, database
```

```python
dataset = pd.read_csv("50_Startups.csv") #read the file data and store it in dataset variable
```

```python
dataset
```

```python
#data preprocessing - because categorial data can't be handle so we modify the data into numerical data using one hot encoding(Nominal)

dataset = pd.get_dummies(dataset, drop_first=True)
```

```python
dataset
```

```python
boolean_columns = dataset.select_dtypes(include=['bool']).columns #get boolean type column data from the dataset and store it in boolean_columns variable
print(boolean_columns)

dataset[boolean_columns] = dataset[boolean_columns].astype(int) #convert boolean column dataset to 0's and 1's
print(dataset[boolean_columns])
```

```python
dataset
```

```python
#get column names

dataset.columns
```

```python
#split input data

independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
independent
```

```python
#split output data

dependent = dataset[['Profit']]
dependent
```

```python
#split the training and testing dataset

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
#Create a model from training set

from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion='poisson', splitter='best')
regressor = regressor.fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
#test the model with test data

y_pred = regressor.predict(X_test)
y_pred
```

```python
#Evaluation metrics

from sklearn.metrics import r2_score
r_score = r2_score(y_test, y_pred) #r_score near to 1 - good model. if near to zero - poor model
r_score
```

```python
#save the model using pickle library

import pickle
fileName = "finalized_model_svr.sav"
pickle.dump(regressor, open(fileName, "wb")) #saved model available in regressor variable. Now storing that variable into fileName
```
